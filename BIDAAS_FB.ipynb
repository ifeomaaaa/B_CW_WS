```python
import pandas as pd
```

```python
df1 = pd.read_csv('/content/facebook.csv')
df1.head()
#ps. reach and reach.1 duplicates
```

```python
df = df1.drop(df1.columns[[0, 2, 11, 12, 13, 15]], axis=1)
df.describe()
```

```python
df.info()
```

```python
max_reach = df[df["reach"] == 33857]
print(max_reach) 
```

```python
min_reach = df[df["reach"] == 1]
print(min_reach) 
```

```python
max_spend = df[df["spend"] == 14201.63]
print(max_spend) 
```

```python
min_spend = df[df["spend"] == 1.72]
print(min_spend) 
```

```python
df[["created_time"]] = df[["created_time"]].apply(pd.to_datetime)
```

```python
df["weekday"] =  df['created_time'].dt.day_name()
df.head(2)
```

```python
df['avg_reach'] = df.groupby('weekday').reach.transform('mean')
df['avg_spend'] = df.groupby('weekday').spend.transform('mean')

df.head(2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.bar(df["weekday"], df["avg_reach"], color="firebrick")
plt.title("Average reach per weekday") 
plt.xlabel("weekday")
plt.ylabel("average reach") 
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.bar(df["weekday"], df["avg_spend"], color="firebrick")
plt.title("Average spend per weekday") 
plt.xlabel("weekday")
plt.ylabel("average spend") 
plt.xticks(rotation = 90)
plt.show()
```
